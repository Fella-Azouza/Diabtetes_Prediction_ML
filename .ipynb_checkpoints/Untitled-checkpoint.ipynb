## Nettoyage et prétraitement des données

```python
# Suppression des lignes où le genre est "other"
df = df[df['gender'] != 'Other']
```

```python
# Afficher les valeurs et leurs fréquences pour chaque colonne
for column in df.columns:
    print(f"Fréquences des valeurs pour {column}:")
    print(df[column].value_counts())
    print("\n")
```

```python
print("Missing values in data set:\n", df.isnull().sum())
```

```python
# Remplacer uniquement les valeurs "No Info" par NaN dans la colonne "smoking_history"
df['smoking_history'] = df['smoking_history'].replace("No Info", np.nan)
```

```python
print(df.isnull().sum())
```

```python
# Calculer la valeur la plus fréquente (mode)
mode_value = df['smoking_history'].mode()[0]

# Remplacer les valeurs NaN par la valeur la plus fréquente
df['smoking_history'] = df['smoking_history'].fillna(mode_value)
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna()  # Supprime les lignes avec des valeurs manquantes
```

## Encoding


#### Encoding Categorical Variables

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

label_encoder = LabelEncoder()



# Liste des colonnes à encoder
categorical_columns = ['gender', 'smoking_history']

for col in categorical_columns:
    # Encoder les valeurs catégorielles
    df[col] = label_encoder.fit_transform(df[col])

print(df.head())
```

## Séparation des features et de la target :

## Verify Independency between features:


```python
correlation_matrix = df.corr()
plt.figure(figsize=(16,8))
print("Matrice de corrélation :\n", correlation_matrix)
sns.heatmap(correlation_matrix,annot=True)
```

```python

# Définir les seuils
target_col = 'diabetes'  # Nom de la colonne cible
target_corr_threshold = 0.1  # Seuil pour la corrélation avec la cible
multicollinearity_threshold = 0.85 # Seuil pour détecter la multicolinéarité



# Étape 1 : Sélectionner les caractéristiques fortement corrélées avec la cible
target_corr = correlation_matrix[target_col].drop(index=target_col)
selected_features = target_corr[abs(target_corr) > target_corr_threshold].index.tolist()

# Étape 2 : Gérer la multicolinéarité
# Créer une matrice de corrélation filtrée pour les caractéristiques sélectionnées
filtered_corr = correlation_matrix.loc[selected_features, selected_features]

# Supprimer une caractéristique de chaque paire hautement corrélée
to_drop = set()
for i in range(len(filtered_corr.columns)):
    for j in range(i):
        if abs(filtered_corr.iloc[i, j]) > multicollinearity_threshold:
            # Comparer la corrélation avec la cible et supprimer la moins importante
            col_i = filtered_corr.columns[i]
            col_j = filtered_corr.columns[j]
            if abs(target_corr[col_i]) < abs(target_corr[col_j]):
                to_drop.add(col_i)
            else:
                to_drop.add(col_j)

# Caractéristiques finales après suppression
final_features = [f for f in selected_features if f not in to_drop]

print("Caractéristiques sélectionnées :", final_features)
```

##  Standardisation des données

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = df[final_features]
# Apprend les paramètres sur l'ensemble d'entraînement
features_scaled = scaler.fit_transform(X)

```

## Split the data

```python
# Diviser les données (80% pour l'entraînement, 20% pour le test)
features_train, features_test, target_train, target_test = train_test_split(features_scaled, target, test_size=0.2, random_state=42)

```

# A-supervised

 ## 1-KNN

```python
k_values = range(1, 21)  # Tester les valeurs de K de 1 à 20
mean_scores = [] 
for k in k_values:
    knn_model = KNeighborsRegressor(n_neighbors=k)
    
    # Utiliser la validation croisée pour calculer l'erreur quadratique moyenne pour chaque K
    scores = cross_val_score(knn_model, features_train, target_train, cv=5, scoring='neg_mean_squared_error')  # 'neg_mean_squared_error' pour MSE
    mean_scores.append(np.mean(scores))  # Calculer la moyenne des scores
```

 ### Tracer la courbe des performances en fonction de K

```python
plt.figure(figsize=(10, 6))
plt.plot(k_values, -np.array(mean_scores), marker='o', linestyle='-', color='b')  # Utilisation de la valeur positive de MSE
plt.xlabel('Valeur de K')
plt.ylabel('Erreur quadratique moyenne (MSE)')
plt.title('Optimisation du paramètre K pour KNN')
plt.show()
```

###  Sélectionner le meilleur K en fonction du score MSE minimum

```python
best_k_knn = k_values[np.argmin(mean_scores)]
print(f"Meilleur K : {best_k_knn}")
```

### Créer un modèle KNN avec le meilleur K

```python
knn_model_best = KNeighborsRegressor(n_neighbors=best_k_knn)
knn_model_best.fit(features_train, target_train)
```

### Prédire sur les données de test

```python
target_pred_knn = knn_model_best.predict(features_test)
```

###  Calculer l'erreur quadratique moyenne (MSE) avec le meilleur K

```python
mse_best = mean_squared_error(target_test, target_pred_knn)
print(f"Mean Squared Error (MSE) avec le meilleur K : {mse_best}")
```

### Calculer le score R2 avec le meilleur K


```python
r2_best = r2_score(target_test, target_pred_knn)
print(f"R-squared (R2 Score) avec le meilleur K : {r2_best}")
```

### Calculer l'exactitude

```python
accuracy_knn = accuracy_score(target_test, target_pred_knn)
print(f"Accuracy (KNN) : {accuracy_knn:.2f}")
```

### Matrice de confusion


```python
conf_matrix = confusion_matrix(target_test, target_pred_knn)
print("\nMatrice de confusion :")
print(conf_matrix)
```

```python
print("Classification Report:\n", classification_report(target_test, target_pred_knn))
```

### Visualisation des résultats

```python
plt.figure(figsize=(10, 6))
plt.scatter(target_test, target_pred_knn, alpha=0.5)
plt.xlabel("Valeurs réelles (target_test)")
plt.ylabel("Valeurs prédites (target_pred_knn)")
plt.title(f"Comparaison des valeurs réelles et prédites (KNN) avec K={best_k_knn}")
plt.show()
```

# Non supervisées

## K-Means

```python
silhouette_scores = []
kmeans_inertias = []
k_range_kmeans = range(2, 11) 
for k in k_range_kmeans:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features_scaled)
    kmeans_inertias.append(kmeans.inertia_)
    labels = kmeans.labels_
    silhouette_avg = silhouette_score(features_scaled, labels)
    silhouette_scores.append(silhouette_avg)
```

```python
best_k = k_range_kmeans[np.argmax(silhouette_scores)]
print(f"Le meilleur K selon l'indice silhouette est : {best_k}")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(k_range_kmeans, kmeans_inertias, marker='o', linestyle='-', color='b')
plt.xlabel('Nombre de clusters (K)')
plt.ylabel('Inertie')
plt.title('Méthode du coude pour K-Means')
plt.show()
```

### Appliquer K-Means avec le meilleur K


```python
kmeans_model = KMeans(n_clusters=best_k, random_state=42)
kmeans_labels = kmeans_model.fit_predict(features_scaled)
silhouette_kmeans = silhouette_score(features_scaled, kmeans_labels)
```

### Récupérer les labels des clusters


```python
final_labels = kmeans.labels_
df['Cluster'] = kmeans_labels
```

### Visualisation des clusters après PCA

```python
import matplotlib.pyplot as plt
pca = PCA(n_components=2)  # Réduction à 2 dimensions
data_reduced = pca.fit_transform(features_scaled)

# Vérifiez que les données réduites (data_reduced) sont prêtes et que la colonne 'Cluster' existe
# data_reduced : Résultat de la réduction de dimension avec PCA (par exemple avec 2 dimensions)
# data['Cluster'] : Colonne contenant les labels des clusters attribués par K-Means

plt.figure(figsize=(10, 6))  # Taille de la figure
plt.scatter(data_reduced[:, 0], data_reduced[:, 1], c=df['Cluster'], cmap='viridis', alpha=0.7)  # Affiche les points des clusters
plt.title('Visualisation des clusters avec K-Means après PCA')  # Titre du graphique
plt.xlabel('Composante principale 1')  # Étiquette de l'axe X
plt.ylabel('Composante principale 2')  # Étiquette de l'axe Y
plt.colorbar(label='Cluster')  # Barre de couleur indiquant les clusters
plt.show()  # Affiche le graphique
```

```python
print(f"Inertie totale : {kmeans.inertia_}")
```

### Calculer les performances de K-Means

```python
final_silhouette_score = silhouette_score(features_scaled, final_labels)
print(f"Score silhouette final (K = {best_k}) : {final_silhouette_score:.2f}")
```
